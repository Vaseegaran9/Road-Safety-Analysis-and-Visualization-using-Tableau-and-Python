```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'C:\\Users\\Dell\\Desktop\\accident data.csv'

try:
    df = pd.read_csv(file_path)
except FileNotFoundError as e:
    print(f"File not found: {e}")
except PermissionError as e:
    print(f"PermissionError: {e}")
```

This line of code converts the 'Accident Date' column in a pandas DataFrame (df) from a string format to a datetime format using the specified date format ("%d-%m-%Y").

```python
df['Accident Date'] = pd.to_datetime(df['Accident Date'], format='%d-%m-%Y')
```

```python
df['Year'] = df['Accident Date'].dt.year
```

```python
unique_years = df['Year'].unique()
```

```python
for year in unique_years:
    # Filter data for the current year
    year_data = df[df['Year'] == year]
    
    # Calculate monthly accidents for the current year
    monthly_accidents = year_data.groupby('Month').size().reset_index(name='Total Accidents')
    
    month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    
    plt.figure(figsize=(10, 6))
    plt.plot(month_names, monthly_accidents['Total Accidents'], marker='o', linestyle='-')
    plt.title(f'Total Accidents by Month - {year}')
    plt.xlabel('Month')
    plt.ylabel('Total Accidents')
    plt.xticks(rotation=45) //rotate xaxis label
    plt.tight_layout() //spacing between subplots
    plt.grid(True)
    plt.show()
```

# Serious casualities by vehicle type

```python
serious_casualties = df[df['Accident_Severity'] == 'Serious']

# Get unique years
unique_years = serious_casualties['Year'].unique()

# Iterate over each year
for year in unique_years:
    # Filter data for the current year
    year_data = serious_casualties[serious_casualties['Year'] == year]
    
    # Calculate serious casualties by vehicle type for the current year
    #year_data.groupby('Vehicle_Type')['Accident_Severity'].count(): This 
    #groups the data in the year_data DataFrame by the 'Vehicle_Type' column 
    #and then counts the number of occurrences of 'Accident_Severity' within each group. 
    #This creates a Series with the 'Vehicle_Type' as the index and the count as the values.
    #.reset_index(): This converts the resulting Series back into a DataFrame. The original 
    #index (which was the 'Vehicle_Type' column) becomes a regular column, and a new default integer index is assigned.
    serious_casualties_by_vehicle = year_data.groupby('Vehicle_Type')['Accident_Severity'].count().reset_index()
    
    print(f"Serious Casualties by Vehicle Type for {year}:")
    print(serious_casualties_by_vehicle)
    print("\n")
```

# Serious casualities by Weather conditions

```python
serious_casualties = df[df['Accident_Severity'] == 'Serious']

# Get unique years
unique_years = serious_casualties['Year'].unique()

# Iterate over each year
for year in unique_years:
    # Filter data for the current year
    year_data = serious_casualties[serious_casualties['Year'] == year]
    
    # Calculate serious casualties by weather condition for the current year
    serious_casualties_by_weather = year_data.groupby('Weather_Conditions')['Accident_Severity'].count().reset_index()
    
    # Create a pie chart
    plt.figure(figsize=(8, 8))
    #autopct='%1.1f%%' one decimal place
    #startangle=140: This argument sets the starting angle at which the pie chart is drawn, 
    #wedgeprops=dict(width=0.4) A smaller value, like 0.4, would result in narrower slices. (cause of donut pie chart)
    plt.pie(serious_casualties_by_weather['Accident_Severity'], labels=serious_casualties_by_weather['Weather_Conditions'], autopct='%1.1f%%', startangle=140,wedgeprops=dict(width=0.4))
    plt.title(f'Serious Casualties by Weather Condition for {year}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

# Serious casualities by Road conditions

```python
serious_casualties = df[df['Accident_Severity'] == 'Serious']

# Get unique years
unique_years = serious_casualties['Year'].unique()

# Iterate over each year
for year in unique_years:
    # Filter data for the current year
    year_data = serious_casualties[serious_casualties['Year'] == year]
    
    # Calculate serious casualties by road surface for the current year
    serious_casualties_by_surface = year_data.groupby('Road_Type')['Accident_Severity'].count().reset_index()
    
    # Sort by the number of casualties in descending order for better visualization
    serious_casualties_by_surface = serious_casualties_by_surface.sort_values(by='Accident_Severity', ascending=False)
    
    # Create a horizontal bar chart
    plt.figure(figsize=(10, 6))
    plt.barh(serious_casualties_by_surface['Road_Type'], serious_casualties_by_surface['Accident_Severity'], color='skyblue')
    
    # Add labels to the right of the bars
    for i, val in enumerate(serious_casualties_by_surface['Accident_Severity']):
        plt.text(val, i, f'{val}', va='center', color='black', fontweight='bold')
    
    plt.title(f'Serious Casualties by Road type for {year}')
    plt.xlabel('Number of Serious Casualties')
    plt.ylabel('Road Type')
    plt.tight_layout()
    plt.grid(True, axis='x', linestyle='--', alpha=0.6)
    plt.show()
```

```python
import pandas as pd
import folium
```

```python
serious_casualties = df[df['Accident_Severity'] == 'Serious']

# Group by latitude and longitude and count the number of serious casualties
serious_casualties_by_location = serious_casualties.groupby(['Latitude', 'Longitude']).size().reset_index(name='Serious Casualties')

unique_years = serious_casualties['Year'].unique()

# Iterate over each year
for year in unique_years:
    # Filter data for the current year
    year_data = serious_casualties[serious_casualties['Year'] == year]

    # Group by latitude and longitude and count the number of serious casualties
    serious_casualties_by_location = year_data.groupby(['Latitude', 'Longitude']).size().reset_index(name='Serious Casualties')

    # Create a base map centered at a specific location
    map = folium.Map(location=[51.509865, -0.118092], zoom_start=10)

    # Add markers for each location with a popup displaying the number of serious casualties
    for index, row in serious_casualties_by_location.iterrows():
        lat = row['Latitude']
        lon = row['Longitude']
        casualties = row['Serious Casualties']

        folium.Marker(
            location=[lat, lon],
            popup=f'Serious Casualties: {casualties}',
            icon=folium.Icon(color='red', icon='info-sign')
        ).add_to(map)

    # Save the map to an HTML file
    map.save(f'C:\\Users\\Dell\\Desktop\\Dv project\\serious\\serious_casualties_map_{year}.html')
```
